```python
def NumberOfCloumns():
    return len(df.columns)
```

```python
def ColToShift(row):
    colCnt = NumberOfCloumns() 
    
    colShift = 0
    for i in range(colCnt-1,0,-1):
        if pd.isnull(row[i]):
            colShift = colShift+1
        else:
            break
    print("column shift: " , colShift)  
    
    return colShift;    
```

```python
df1 = df
#df1 = df[12:16]
#df1 = df[4:6]
```

```python
colEndInd = NumberOfCloumns()-1 
for index, row in df1.iterrows():   
    print(index, " Before Change " , row['sno'], row['date'], row['desc'],row['iref'], row['amount'], row['tranType'],row['balance'] )            
    
        
    # Column of a row to be shifted towards right
    colShift = ColToShift(row)    
        
    if colShift >0:        
        for i in range(colEndInd,1,-1):                       
            colNum = i-colShift  
            
            print("Colums Index: " , i, "shiftPos", colNum)              
            
            if colNum <= 0:
                row[i]="" # empty, since no value available
            else:
                row[i] = row[colNum]
    
    # Moving fixed columns 
    row['date'] =  row['sno'] 
    row['sno'] = index + 1 
    
    # Persist the data type 
    row['balance'] = str(row['balance'])    
    
    print(index, " after Change " , row['sno'], row['date'], row['desc'],row['iref'], row['amount'], row['tranType'],row['balance'] )            
    print()
    print(type(row.balance))
    df1.loc[index] = row    
```
